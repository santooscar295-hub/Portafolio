## Step 1: Importing Required Libraries

As the name suggests, the first step is to import all the necessary libraries that will be used in the project. In this case, we need pandas, numpy, matplotlib, seaborn, and random libraries.

Pandas is a library used for data manipulation and analysis. Numpy is a library used for numerical computations. Matplotlib is a library used for data visualization. Seaborn is a library used for statistical data visualization. Random is a library used to generate random numbers.

```python
# Importing essential libraries for data analysis and visualization:
# - pandas (pd) for data manipulation and analysis
# - numpy (np) for numerical computations
# - matplotlib.pyplot (plt) for creating plots and charts
# - seaborn (sns) for advanced statistical visualizations
# - random for generating random data or selections
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random

# Creating a sample dataset of posts:
# - 'titulos' generates 20 post titles (post 0 to post 19)
# - 'categorias' defines possible categories for posts
# - 'categorias_random' randomly assigns a category to each post
# - 'likes' generates a random number of likes (10 to 500) for each post
# - A DataFrame 'df' is created with columns Title, Category, and Likes
# - print(df) displays the dataset
titulos = [f"post {i}" for i in range(1, 21)]

categorias = ["Motivation", "Comedy", "Games", "News", "Sports"]

categorias_random = [random.choice(categorias) for _ in range(20)]

likes = [random.randint(10, 500) for _ in range(20)]

df = pd.DataFrame({
    "Title": titulos,
    "Category": categorias_random,
    "Likes": likes
    
})

print(df)



```

```python
#df.info() shows the structure of the DataFrame, including column names, data types, and number of non-null entries.#
df.info()
df
```

```python
# This code generates a bar chart showing the average Likes for each Category in the DataFrame:
# - plt.figure(figsize=(8,5)) sets the size of the plot.
# - sns.barplot(...) creates a bar plot with Categories on the x-axis and Likes on the y-axis.
# - ci=None disables the confidence interval shading.
# - plt.title, plt.ylabel, plt.xlabel add a title and axis labels.
# - plt.show() displays the plot.
plt.figure(figsize=(8,5))
sns.barplot(x="Category", y="Likes", data=df, ci=None)
plt.title("Likes promedio por categoría")
plt.ylabel("Likes")
plt.xlabel("Categoría")
plt.show()
```

```python
# This code generates a horizontal bar chart showing the average Likes for each Category:
# - plt.figure(figsize=(8,5)) sets the figure size.
# - sns.barplot(x="Likes", y="Category", data=df, ci=None) creates a horizontal bar plot.
# - ci=None disables confidence intervals.
# - plt.title adds a title, plt.ylabel and plt.xlabel label the axes.
# - plt.show() displays the plot.
plt.figure(figsize=(8,5))
sns.barplot(x="Likes", y="Category", data=df, ci=None)
plt.title("Likes promedio por categoría")
plt.ylabel("categoria")
plt.xlabel("Likes")
plt.show()
```

```python
# This code generates a bar chart showing average Likes per Category, sorted from highest to lowest:
# - plt.figure(figsize=(8,5)) sets the figure size.
# - sns.barplot(x=promedios.index, y=promedios.values, palette="viridis") plots the sorted averages with a color palette.
# - plt.title, plt.ylabel, plt.xlabel add a descriptive title and axis labels.
# - plt.show() displays the chart.
plt.figure(figsize=(8,5))
sns.barplot(x=promedios.index, y=promedios.values, palette="viridis")
plt.title("Likes promedio por categoría (de mayor a menor)")
plt.ylabel("Likes promedio")
plt.xlabel("Categoría")
plt.show()
```

```python
# The following code sorts the DataFrame in descending order by Likes and removes duplicate Likes:
# - df.sort_values("Likes", ascending=False) sorts posts so that the highest Likes come first.
# - df.drop_duplicates(subset="Likes", keep="first") removes duplicate Likes, keeping the first occurrence (highest Like).
# - df displays the cleaned DataFrame.
df = df.sort_values("Likes", ascending=False)
df = df.drop_duplicates(subset="Likes", keep="first")
df
```

```python
# This code calculates the average Likes per Category and sorts them in descending order:
# - df.groupby("Category")["Likes"].mean() computes the mean Likes for each Category.
# - sort_values(ascending=False) sorts categories from highest to lowest average Likes.
# - reset_index() converts the result back to a DataFrame with a proper index.
# - round(2) rounds the average Likes to two decimal places for readability.
# - tabla_promedio displays the final summarized table.
tabla_promedio = df.groupby("Category")["Likes"].mean().sort_values(ascending=False).reset_index()
tabla_promedio["Likes"] = tabla_promedio["Likes"].round(2)

tabla_promedio
```
